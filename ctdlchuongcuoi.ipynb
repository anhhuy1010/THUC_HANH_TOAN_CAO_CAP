```python
class Node: # tạo lớp node
  def __init__(self, data): # tạo hàm khởi tạo 
    self.left = None # tạo hàm bên trái bằng 0
    self.right = None # tạo hàm bên phải bằng
    self.data = data # gán giá trị data bằng data
#Insert to create nodes
  def insert(self, data): # tạo hàm insert
    if self.data: # nếu self
      if data < self.data: # nếu data bé hớn self.data
        if self.left is None: # nếu self.left bằng 0
          self.left = Node(data) # gán bên trái bằng data
        else: # ngược lại
          self.left.insert(data) # thêm giá trị vào bên trái
      if data > self.data: # nếu data lớn hơn self.data
        if self.right is None: # nếu bên phải bằng 0
          self.right = Node(data) # bên trái bằng data
        else: # ngược lại 
          self.right.insert(data) # thêm giá trị vào bên phải
    else: # ngược lại
      self.data = data # data bằng self.data
#find value to compare the value
  def findval(self, inval): # tạo hàm tìm kiếm 
    if (inval < self.data): # neeusinval bé hơn self.data
      if self.left is None: # nếu bên trái bằng 0
        return (str(inval)+" Not Found") # trả lại giá trị  inval 
      return (self.left.findval(inval)) # trả lại giá trị bên trái
    if (inval > self.data): # nếu inval lớn hơn self.data
      if self.right is None: # nếu bên phải bằng 0
        return (str(inval)+" Not Found") # trả gái trị inval 
      return self.right.findval(inval) # trả về giá trị bên phải
    else: # ngược lại 
      print(str(self.data) + ' is found') # in ra giá trị của self.data
  def PrintTree(self): # khởi tạo hàm printtree
    if self.left: # nếu gái trị bên trái
      self.left.PrintTree() #gán vào bên trái printtree
      print( self.data) # in ra self.data
    if self.right: # nếu gái trị bên phải
      self.right.PrintTree() # gán vào bên phải hàm printtree
      print( self.data) # in ra self.data

root = Node(15) #gán giá trị

root.insert(10)#gán giá trị

root.insert(19)#gán giá trị

root.insert(5)#gán giá trị
root.insert(13)#gán giá trị
root.insert(14)#gán giá trị
root.insert(29)#gán giá trị

print(root.findval(29))#gán giá trị
print(root.findval(10))#gán giá trị
print(root.findval(40))#gán giá trị
```
